### Вопрос 6
Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю

```python
# Вопрос 6: Необходимо соединить все фреймы, содать колонки, овечающие за RFM и удалить лишние
```

```python
# Пошагово создаем фрейм для rfm
r = for_rfm.groupby('customer_unique_id', as_index=False).agg({'order_purchase_timestamp':'max'})
```

```python
f = for_rfm.groupby('customer_unique_id', as_index=False).agg({'customer_id':'count'})
```

```python
m = for_rfm.groupby('customer_unique_id', as_index=False).agg({'price':'sum'})
```

```python
rf = r.merge(f, on='customer_unique_id')
```

```python
rfm = rf.merge(m, on='customer_unique_id')
```

```python
current_date = datetime.now()
```

```python
rfm['order_purchase_timestamp'] = rfm['order_purchase_timestamp'].apply(lambda x: (current_date - x).days)
```

```python
rfm = rfm.rename(columns={'order_purchase_timestamp':'recency_days', 'customer_id':'frequency_number'})
```

```python
rfm
```

```python
# Посмотрим на границы данных, которые потом будут определять границы кластеров
rfm.describe()
```

```python
# Устанавливаем границы кластеров через квантили
quintiles = rfm[['recency_days', 'price']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
quintiles['frequency_number'] = {0.2: 1.0, 0.4: 5.0, 0.6: 10.0, 0.8: 20.0}
```

```python
quintiles
```

```python
# Создание функций для распределения по кластерам
def r_score(x):
    if x <= quintiles['recency_days'][.2]:
        return 5
    elif x <= quintiles['recency_days'][.4]:
        return 4
    elif x <= quintiles['recency_days'][.6]:
        return 3
    elif x <= quintiles['recency_days'][.8]:
        return 2
    else:
        return 1

def fm_score(x, c):
    if x <= quintiles[c][.2]:
        return 1
    elif x <= quintiles[c][.4]:
        return 2
    elif x <= quintiles[c][.6]:
        return 3
    elif x <= quintiles[c][.8]:
        return 4
    else:
        return 5
```

```python
# Создание новых колонок и распределение данных по кластерам
rfm['R'] = rfm['recency_days'].apply(lambda x: r_score(x))
rfm['F'] = rfm['frequency_number'].apply(lambda x: fm_score(x, 'frequency_number'))
rfm['M'] = rfm['price'].apply(lambda x: fm_score(x, 'price'))
```
